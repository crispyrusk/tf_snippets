```python
import tensorflow as tf
```

```python
names = ['test' + str(x) for x in range(0, 100)]
print(names)
```

```python
# setup a queue
queue = tf.train.input_producer(names, shuffle=False)
```

```python
front = queue.dequeue()
front_many = queue.dequeue_many(n=5)
```

https://www.tensorflow.org/programmers_guide/supervisor

```python
queue_size = queue.size()
```

```python
tf.summary.histogram("queue_size", queue_size)
```

```python
my_summary_op = tf.summary.merge_all()
```

```python
# turn off automatic logging and explicity control logging
# example: https://www.tensorflow.org/programmers_guide/supervisor#larger_model_scenario
sv = tf.train.Supervisor(logdir="logsupervisor", summary_op=None)
```

```python
with sv.managed_session() as sess:
    for step in range(1000):
        if sv.should_stop():
            break
    
        ff, summ = sess.run([front, my_summary_op]) # compute the summary as a part of the main thread
        sv.summary_computed(sess, summ) 
        print(ff)
```
