# Constants vs Variables

* Constant values are stored in the graph definition
* Sessions allocate memory to store variable values

```python
import tensorflow as tf
```

```python
a = tf.constant(10)
b = tf.constant(20)
```

```python
c = a + b
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    # not the values of the constant are stored in the graph
    print(sess.graph_def)
    sess.run(c)
    
    
    
```

```python
g = tf.Graph()
with g.as_default():
    x_a = tf.Variable(10)
    x_b = tf.Variable(20)
    x_c = x_a + x_b
    init2 = tf.global_variables_initializer()
    print(type(x_a))
    print(type(x_b))
    print(type(x_c))
    
    with tf.Session() as sess:
        sess.run(init2)
        
        # note that session allocates memory to store variables
        print(sess.graph_def)
```
