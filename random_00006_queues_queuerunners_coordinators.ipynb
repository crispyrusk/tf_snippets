```python
import tensorflow as tf
```

```python
names = ['test' + str(x) for x in range(0, 100)]
```

```python
queue = tf.train.input_producer(names, shuffle=False)
```

```python
front = queue.dequeue()
front_many = queue.dequeue_many(n=5)
```

```python
init_op = tf.global_variables_initializer()
```

** Queue Runner are responsible for enqueueing **

** and Coordinator is responsible for stop threads safely**

```python
with tf.Session() as sess:
    sess.run(init_op)

    # build a coord to coordinate all the child threads in the main thread
    # coord orchestrates shutting down of all threads
    coord = tf.train.Coordinator()
    
    # initialize the queue threads and lets the coordinator take control of the threads
    # Beware, if you don't start all your queues before runnig anything
    # The main threads will wait for them to start and you will hang again
    # This helper start all queues in tf.GraphKeys.QUEUE_RUNNERS
    # This helper will associate a thread with every queue-runner, which will try to keep the queues non-empty
    threads = tf.train.start_queue_runners(coord=coord)
    
    for i in range(0, 200):
        print(sess.run(front))
        print(sess.run(front_many))
    
    # note that even though we have asked for too many dequeue operations, the threads wont hang. 
    # if threads are done, they go back to main threads
    
    # request all threads to stop
    coord.request_stop()
    
    # wait for all threads to join
    coord.join(threads)
    
```
