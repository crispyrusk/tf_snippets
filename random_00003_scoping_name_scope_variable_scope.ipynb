get_variable plays well only with variable_scope.
It doesnt respect name_scope

```python
with tf.name_scope("namescope"):
    a = tf.constant(1, name="a")
    b = tf.constant(2, name="b")
    bb = tf.constant(3, name="b") # note this varialbe gets renamed as namescope/b_1 - no expection raised; but get_variable would have
    
    # this is the maing point, "c" is not a part of the "namescope"
    c = tf.get_variable(name="c", shape=[1,2], dtype=tf.int32)
```

```python
with tf.variable_scope("varscope") as scope:
    a = tf.constant(10, name="a")
    b = tf.constant(20, name="b")
    c = tf.get_variable(name="c", shape=[10,20], dtype=tf.int32)
    #d = tf.get_variable(name="c", shape=[10,20], dtype=tf.int32) # this will cause error because vars are not reused
    scope.reuse_variables()
    d = tf.get_variable(name="c", shape=[10,20], dtype=tf.int32) # this is a shared variable
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    print(sess.graph_def)
```
